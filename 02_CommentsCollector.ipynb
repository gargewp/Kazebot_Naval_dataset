```python
# Собираем и сохраняем комменты с одного видео
def save_comment_list(video_Id, author_video_name, storage_dir_name, api_key):
    record_strings = scanning_lib.scan_single_video(video_Id, author_video_name, api_key)

    if not os.path.isdir(storage_dir_name):
        os.mkdir(storage_dir_name)
        
    df_comments = pd.DataFrame(data=record_strings, columns=column_names_comment)
    comment_file_name = storage_dir_name + """/""" + video_Id + "_video_comments.csv"
    
    df_comments.to_csv(comment_file_name)
    return 0
```

```python
# Сохраняем комменты для каждого видео в отдельный файл
for api_key in api_keys:
    
    df_fvideos = pd.read_csv(videos_table)[column_names_video]
    
    comment_files = [storage_dir_name + """/""" + file_name for file_name in os.listdir(storage_dir_name) 
                     if file_name.find("_video_comments.csv") >= 0]
    existed_v_ids = [re.split(storage_dir_name + r'\/(.*)_video_comments.csv', comment_file)[1] 
                             for comment_file in comment_files]
    
    try: 
        print("Start using new key")
        for record in df_fvideos.T.iteritems():
            video_Id = record[1]["id"]
            ans = "Already scanned"
            if not any([existed_v_id == video_Id for existed_v_id in existed_v_ids]):
                author_video_name = record[1]["channelTitle"]
                ans = save_comment_list(video_Id, author_video_name, storage_dir_name, api_key)
            print(ans)
        break
    except:
        print("Change key")
        continue
print("Collection completed")
```
